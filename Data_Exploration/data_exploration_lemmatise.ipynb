```python
import stanza

nlp = stanza.Pipeline('ar', use_gpu=True, download_method=None)

def pickle_lemma(text, out_path):
    # Split into chunks of ~1000 words
    words = text.split()
    chunk_size = 1000
    chunks = [" ".join(words[i:i+chunk_size]) for i in range(0, len(words), chunk_size)]
    
    lemma = []
    for chunk in chunks:
        doc = nlp(chunk)
        lemma += [word.lemma for sentence in doc.sentences for word in sentence.words]
    
    with open(out_path, 'wb') as f:
        pickle.dump(lemma, f)

# pickle_lemma(txt_1001, 'pickles_lemma/1001_lemma.pkl')
# pickle_lemma(txt_ALC, 'pickles_lemma/ALC_lemma.pkl')
# pickle_lemma(txt_BTEC, 'pickles_lemma/BTEC_lemma.pkl')
# pickle_lemma(txt_Hadith, 'pickles_lemma/Hadith_lemma.pkl')
# pickle_lemma(txt_Hayy, 'pickles_lemma/Hayy_lemma.pkl')
# pickle_lemma(txt_NT, 'pickles_lemma/NT_lemma.pkl')
# pickle_lemma(txt_Odes, 'pickles_lemma/Odes_lemma.pkl')
# pickle_lemma(txt_OT, 'pickles_lemma/OT_lemma.pkl')
# pickle_lemma(txt_QALB, 'pickles_lemma/QALB_lemma.pkl')
# pickle_lemma(txt_Quran, 'pickles_lemma/Quran_lemma.pkl')
# pickle_lemma(txt_Sara, 'pickles_lemma/Sara_lemma.pkl')
# pickle_lemma(txt_Wikinews, 'pickles_lemma/Wikinews_lemma.pkl')
# pickle_lemma(txt_ZAEBUC, 'pickles_lemma/ZAEBUC_lemma.pkl')
```

```python
import re

def remove_diacritics(text):
    return re.sub(r'[\u0610-\u061A\u064B-\u065F\u0670]', '', text)

def clean_and_repickle(input_path, output_path):
    # Load existing pickle
    with open(input_path, 'rb') as f:
        lemma_list = pickle.load(f)
    
    # Remove diacritics from each lemma
    cleaned = [remove_diacritics(lemma) for lemma in lemma_list]
    
    # Save cleaned lemmas
    with open(output_path, 'wb') as f:
        pickle.dump(cleaned, f)
    
    print(f"Done: {input_path} → {output_path} ({len(cleaned)} lemmas)")

# Run for all corpora
# clean_and_repickle('pickles_lemma/1001_lemma.pkl',     'pickles_lemma/1001_lemma.pkl')
# clean_and_repickle('pickles_lemma/ALC_lemma.pkl',      'pickles_lemma/ALC_lemma.pkl')
# clean_and_repickle('pickles_lemma/BTEC_lemma.pkl',     'pickles_lemma/BTEC_lemma.pkl')
# clean_and_repickle('pickles_lemma/Hadith_lemma.pkl',   'pickles_lemma/Hadith_lemma.pkl')
# clean_and_repickle('pickles_lemma/Hayy_lemma.pkl',     'pickles_lemma/Hayy_lemma.pkl')
# clean_and_repickle('pickles_lemma/NT_lemma.pkl',       'pickles_lemma/NT_lemma.pkl')
# clean_and_repickle('pickles_lemma/Odes_lemma.pkl',     'pickles_lemma/Odes_lemma.pkl')
# clean_and_repickle('pickles_lemma/OT_lemma.pkl',       'pickles_lemma/OT_lemma.pkl')
# clean_and_repickle('pickles_lemma/QALB_lemma.pkl',     'pickles_lemma/QALB_lemma.pkl')
# clean_and_repickle('pickles_lemma/Quran_lemma.pkl',    'pickles_lemma/Quran_lemma.pkl')
# clean_and_repickle('pickles_lemma/Sara_lemma.pkl',     'pickles_lemma/Sara_lemma.pkl')
# clean_and_repickle('pickles_lemma/Wikinews_lemma.pkl', 'pickles_lemma/Wikinews_lemma.pkl')
# clean_and_repickle('pickles_lemma/ZAEBUC_lemma.pkl',   'pickles_lemma/ZAEBUC_lemma.pkl')
# lemma_1001 = load_from_pickle('pickles_lemma/1001_lemma.pkl')
# lemma_ALC = load_from_pickle('pickles_lemma/ALC_lemma.pkl')
# lemma_BTEC = load_from_pickle('pickles_lemma/BTEC_lemma.pkl')
# lemma_Hadith = load_from_pickle('pickles_lemma/Hadith_lemma.pkl')
# lemma_Hayy = load_from_pickle('pickles_lemma/Hayy_lemma.pkl')
# lemma_NT = load_from_pickle('pickles_lemma/NT_lemma.pkl')
# lemma_Odes = load_from_pickle('pickles_lemma/Odes_lemma.pkl')
# lemma_OT = load_from_pickle('pickles_lemma/OT_lemma.pkl')
# lemma_QALB = load_from_pickle('pickles_lemma/QALB_lemma.pkl')
# lemma_Quran = load_from_pickle('pickles_lemma/Quran_lemma.pkl')
# lemma_Sara = load_from_pickle('pickles_lemma/Sara_lemma.pkl')
# lemma_Wikinews = load_from_pickle('pickles_lemma/Wikinews_lemma.pkl')
# lemma_ZAEBUC = load_from_pickle('pickles_lemma/ZAEBUC_lemma.pkl')
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import arabic_reshaper
from bidi.algorithm import get_display

def fix_arabic(text):
    reshaped = arabic_reshaper.reshape(text)
    return get_display(reshaped)

def plot_top_words(lemma_list, title, save_image_path, stopwords_removal=False):
    if stopwords_removal:
        lemma_list = [lemma for lemma in lemma_list if lemma not in arabic_stopwords]
    
    lemma_counts = Counter(lemma_list)
    
    top_20 = lemma_counts.most_common(20)
    print(f"Top 20 lemmas in {title} (Stopwords {'Removed' if stopwords_removal else 'Included'}):")
    for lemma, count in top_20:
        print(f"{lemma}: {count}")
    
    words, counts = zip(*top_20)
    words = [fix_arabic(w) for w in words] 
    
    plt.figure(figsize=(10, 6))
    plt.bar(words, counts)
    plt.title(f"Top 20 Lemmas in {title} (Stopwords {'Removed' if stopwords_removal else 'Included'})")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(save_image_path, dpi=500)
    plt.close()
    return lemma_list
```

```python
def pickle_cleaned_lemma(lemma_list, out_path):
    with open(out_path, 'wb') as f:
        pickle.dump(lemma_list, f)
```

```python
import os
import re

def combine_sent_files(folder_path, output_file):
    def natural_sort_key(s):
        return [int(text) if text.isdigit() else text.lower()
                for text in re.split(r'(\d+)', s)]

    # Get only .sent files
    files = [f for f in os.listdir(folder_path) if f.endswith(".sent")]

    # Sort naturally
    files.sort(key=natural_sort_key)

    print("Files will be combined in this order:\n")

    with open(output_file, "w", encoding="utf-8") as outfile:
        for i, filename in enumerate(files, start=1):
            print(f"{i:03d} -> {filename}")
            
            file_path = os.path.join(folder_path, filename)
            
            with open(file_path, "r", encoding="utf-8") as infile:
                content = infile.read().strip()
                
                outfile.write(content)
                outfile.write("\n")

    print("\nAll files combined successfully.")
```

```python
in_1001 = r"Corpora\camel_treebank_1.1\camel_treebank_1.1\data\sent\1001"
out_1001 = r"Final_Corpus\1001_combined.txt"

combine_sent_files(in_1001, out_1001)
```

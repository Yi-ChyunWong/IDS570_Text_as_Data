```python
import json
from gensim.models import Word2Vec

def train_word2vec(corpus_file, model_file):
    with open(corpus_file, 'r', encoding='utf-8') as f:
        corpus = json.load(f)
    
    docs = []
    for chapter_id, sentences in corpus.items():
        for sentence in sentences:
            lemmas = sentence['lemmas']
            docs.append(lemmas)
    
    model = Word2Vec(sentences=docs, vector_size=200, window=5, min_count=5, workers=20, sg=1)
    model.save(model_file)

train_word2vec('corpus_ner.json', 'models/word2vec_corpus_all.bin')
```

```python
from gensim.models import Word2Vec

def print_similar_words(model_file, word, topn=30):
    model = Word2Vec.load(model_file)
    if word in model.wv:
        similar_words = model.wv.most_similar(word, topn=topn)
        print(f"Words similar to '{word}':")
        for similar_word, similarity in similar_words:
            print(f"{similar_word}: {similarity:.4f}")
    else:
        print(f"'{word}' not found in the model vocabulary.")

power_words = ["الملك", "ملك", "سلطان", "أمير", "خليفة", "الخليفة", "عسكر"]
for word in power_words:
    print_similar_words('models/word2vec_corpus_all.bin', word)
    print()
```

```python
religion_words = ["الله", "الدين", "دين", "مسلم", "مؤمن", "كفار", "إسلام"]
for word in religion_words:
    print_similar_words('models/word2vec_corpus_all.bin', word)
    print()
```

```python
import json

def find_sentences_with_word(corpus_file, target_word):
    with open(corpus_file, 'r', encoding='utf-8') as f:
        corpus = json.load(f)
    
    sentences_with_word = []
    for chapter_id, sentences in corpus.items():
        for sentence in sentences:
            lemmas = sentence['lemmas']
            if target_word in lemmas:
                sentences_with_word.append(sentence['sentence'])
    
    return sentences_with_word

# sentences_with_word = find_sentences_with_word('corpus_ner.json', 'الدين')
# print(f"Sentences containing 'دين':")
# for sentence in sentences_with_word:
#     print(sentence)
#     print("Done")
#     print()


for sent in find_sentences_with_word('corpus_ner.json', 'السعيد'):
    print(sent)
    print("Done")
    print()
```

```python
Power_A = ["الملك", "ملك", "سلطان", "أمير", "خليفة", "الخليفة", "عسكر"]
Power_B = ["تاج", "ملوك", "شاه", "ملكة", "دولة", "عساكر", "جيش", "جيوش"]
Power_all = Power_A + Power_B


Religion_A = ["الله", "الدين", "دين", "مسلم", "مؤمن", "كفار", "إسلام"]
Religion_B = ["حمد", "شيخ", "دير"]
Religion_all = Religion_A + Religion_B
```

```python
def label_classes(in_file, out_file, class_words):
    with open(in_file, 'r', encoding='utf-8') as f:
        corpus = json.load(f)
    
    class_0 = 0
    class_1 = 0
    for chapter_id, sentences in corpus.items():
        for sentence in sentences:
            lemmas = sentence['lemmas']
            label = 0
            for word in lemmas:
                if word in class_words:
                    label = 1
                    class_1 += 1
                    break
            if label == 0:
                class_0 += 1
            sentence['label'] = label
    print(f"Class 0 (not in class_words): {class_0} sentences")
    print(f"Class 1 (in class_words): {class_1} sentences")
    print()
    
    with open(out_file, 'w', encoding='utf-8') as f:
        json.dump(corpus, f, ensure_ascii=False, indent=4)

print("Labeling Power class...")
label_classes('corpus_ner.json', 'corpus_labeled_power.json', Power_all)

print("Labeling Religion class...")
label_classes('corpus_ner.json', 'corpus_labeled_religion.json', Religion_all)
```

```python
import json
import random

def split_data(in_file, out_train_file, out_test_file, test_size=0.2):
    random.seed(42)
    with open(in_file, 'r', encoding='utf-8') as f:
        corpus = json.load(f)
    
    class_0 = []
    class_1 = []
    for chapter_id, sentences in corpus.items():
        for sentence in sentences:
            if sentence['label'] == 0:
                class_0.append((' '.join(sentence['lemmas']), 0))
            else:
                class_1.append((' '.join(sentence['lemmas']), 1))
    
    class_0_sample = random.sample(class_0, len(class_1))
    combined = class_0_sample + class_1
    random.shuffle(combined)
    split_index = int(len(combined) * (1 - test_size))
    train_data = combined[:split_index]
    test_data = combined[split_index:]

    with open(out_train_file, 'w', encoding='utf-8') as f:
        json.dump(train_data, f, ensure_ascii=False, indent=4)

    with open(out_test_file, 'w', encoding='utf-8') as f:
        json.dump(test_data, f, ensure_ascii=False, indent=4)

split_data('corpus_labeled_power.json', 'train_power.json', 'test_power.json')
split_data('corpus_labeled_religion.json', 'train_religion.json', 'test_religion.json')
```

```python
import json
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
import joblib

def train_classifier(train_file, test_file, penalty, title):
    with open(train_file, 'r', encoding='utf-8') as f:
        train_data = json.load(f)
    with open(test_file, 'r', encoding='utf-8') as f:
        test_data = json.load(f)

    X_train = [lemmas for (lemmas, label) in train_data]
    y_train = [label for (lemmas, label) in train_data]
    X_test = [lemmas for (lemmas, label) in test_data]
    y_test = [label for (lemmas, label) in test_data]

    vectorizer = TfidfVectorizer()
    X_train_tfidf = vectorizer.fit_transform(X_train)
    X_test_tfidf = vectorizer.transform(X_test)

    print(X_train_tfidf.shape)
    print(X_test_tfidf.shape)

    l1_r = 1 if penalty == 'l1' else 0
    solver = 'liblinear' if penalty == 'l1' else 'lbfgs'
    clf = LogisticRegression(max_iter=1000, n_jobs=20, l1_ratio=l1_r, solver=solver)
    clf.fit(X_train_tfidf, y_train)
    y_pred = clf.predict(X_test_tfidf)
    y_prob = clf.predict_proba(X_test_tfidf)[:, 1]

    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:")
    print(cm)

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    auc = roc_auc_score(y_test, y_prob)
    print("ROC AUC:", round(auc, 3))

    print("\nNumber of non-zero coefficients: ", (clf.coef_ != 0).sum())
    top_pos_words = sorted(
        zip(vectorizer.get_feature_names_out(), clf.coef_[0]),
        key=lambda x: x[1],
        reverse=True
    )[:20]
    print("\nTop positive words:")
    for word, coef in top_pos_words:
        print(f"{word}: {coef:.3f}")

    top_neg_words = sorted(
        zip(vectorizer.get_feature_names_out(), clf.coef_[0]),
        key=lambda x: x[1],
        reverse=False
    )[:20]
    print("\nTop negative words:")
    for word, coef in top_neg_words:
        print(f"{word}: {coef:.3f}")


    joblib.dump(vectorizer, f"models/tfidf_vectorizer_{title}_{penalty}.joblib")
    joblib.dump(clf, f"models/logistic_regression_classifier_{title}_{penalty}.joblib")

print("Training Power classifier L1...")
train_classifier('train_power.json', 'test_power.json', 'l1', 'power')
print()
print("Training Power classifier L2...")
train_classifier('train_power.json', 'test_power.json', 'l2', 'power')
print()
```

```python
print("Training Religion classifier L1...")
train_classifier('train_religion.json', 'test_religion.json', 'l1', 'religion')
print()
print("Training Religion classifier L2...")
train_classifier('train_religion.json', 'test_religion.json', 'l2', 'religion')
```

```python
import stanza

nlp = stanza.Pipeline(
    lang='ar',
    dir='C:/Users/yichy/stanza_resources',
    download_method=None,
    use_gpu=True
)
```

```python
import json

def corpus_ner_2():
    with open('corpus_sentences.json', 'r', encoding='utf-8') as f:
        corpus = json.load(f)

    for chapter_id, sentence_list in corpus.items():

        for sentence in sentence_list:
            lemmas = sentence['lemmas']
            sentence_text = " ".join(lemmas)
            doc = nlp(sentence_text)
            entities = []
            for ent in doc.ents:
                entities.append({
                    'text': ent.text,
                    'type': ent.type,
                })
                    
            sentence['entities'] = entities
            print(entities)
        print(f"Processed {chapter_id}")

    with open('corpus_ner_2.json', 'w', encoding='utf-8') as f:
        json.dump(corpus, f, ensure_ascii=False, indent=4)

corpus_ner_2()
```

```python
import json

def corpus_ner():
    with open('corpus_sentences.json', 'r', encoding='utf-8') as f:
        corpus = json.load(f)

    for chapter_id, sentence_list in corpus.items():

        for sentence in sentence_list:
            sentence_text = sentence['sentence']
            doc = nlp(sentence_text)
            entities = []
            for ent in doc.ents:
                entities.append({
                    'text': ent.text,
                    'type': ent.type,
                })
                    
            sentence['entities'] = entities
            print(entities)
        print(f"Processed {chapter_id}")

    with open('corpus_ner.json', 'w', encoding='utf-8') as f:
        json.dump(corpus, f, ensure_ascii=False, indent=4)

corpus_ner()
```

```python
from collections import Counter
import json

def count_entity_types():
    with open('corpus_ner.json', 'r', encoding='utf-8') as f:
        corpus = json.load(f)

    type_counts = Counter()

    for chapter_id, sentence_list in corpus.items():
        for sentence in sentence_list:
            for ent in sentence.get('entities', []):
                type_counts[ent['type']] += 1

    return type_counts

entity_type_counts = count_entity_types()
for ent_type, count in entity_type_counts.most_common():
    print(f"{ent_type}: {count}")
```

```python
from collections import defaultdict
import json

def cooccurence(word_list, in_file, out_file):
    with open(in_file, 'r', encoding='utf-8') as f:
        corpus = json.load(f)

    filtered_corpus = {}

    for chapter_id, sentence_list in corpus.items():
        print(f"Processing {chapter_id}...")
        filtered_sentences = []

        for sentence in sentence_list:
            lemmas = sentence.get('lemmas', [])
            entities = sentence.get('entities', [])

            has_power = any(word in word_list for word in lemmas)
            has_entity = len(entities) > 0

            if has_power and has_entity:
                filtered_sentences.append(sentence)
        
        filtered_corpus[chapter_id] = filtered_sentences
    
    with open(out_file, 'w', encoding='utf-8') as f:
        json.dump(filtered_corpus, f, ensure_ascii=False, indent=4)
        

power_words = ["الملك", "ملك", "سلطان", "أمير", "خليفة", "الخليفة", "عسكر"]
religion_words = ["الله", "الدين", "دين", "مسلم", "مؤمن", "كفار", "إسلام"]
cooccurence(power_words, 'corpus_ner.json', 'corpus_cooccurence.json')
cooccurence(religion_words, 'corpus_ner.json', 'corpus_religion.json')
```

```python
from collections import Counter
import json

def count_entities(in_file):
    with open(in_file, 'r', encoding='utf-8') as f:
        corpus = json.load(f)

    entity_counts = Counter()

    for chapter_id, sentence_list in corpus.items():
        for sentence in sentence_list:
            for ent in sentence.get('entities', []):
                entity_counts[ent['text']] += 1

    return entity_counts

entity_counts = count_entities('corpus_cooccurence.json')
for ent_text, count in entity_counts.most_common():
    print(f"{ent_text}: {count}")
```

```python
import json

def split_json_by_labels(input_file, output_file_label0, output_file_label1):
    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    label_0_data = {}
    label_1_data = {}


    for chapter_id, sentence_list in data.items():
        label_0_sentences = []
        label_1_sentences = []

        for sentence_obj in sentence_list:
            sentence = sentence_obj.get("sentence")
            label = sentence_obj.get("label")

            if sentence is None or label is None:
                continue

            reduced_entry = {
                "sentence": sentence,
                "label": label
            }

            if label == 0:
                label_0_sentences.append(reduced_entry)
            elif label == 1:
                label_1_sentences.append(reduced_entry)

        label_0_data[chapter_id] = label_0_sentences
        label_1_data[chapter_id] = label_1_sentences

    with open(output_file_label0, 'w', encoding='utf-8') as f:
        json.dump(label_0_data, f, ensure_ascii=False, indent=4)

    with open(output_file_label1, 'w', encoding='utf-8') as f:
        json.dump(label_1_data, f, ensure_ascii=False, indent=4)


split_json_by_labels(
    input_file="corpus_labeled_power.json",
    output_file_label0="split_labels/power_label_0.json",
    output_file_label1="split_labels/power_label_1.json"
)

split_json_by_labels(
    input_file="corpus_labeled_religion.json",
    output_file_label0="split_labels/religion_label_0.json",
    output_file_label1="split_labels/religion_label_1.json"
)
```

```python
import json

with open("split_labels/power_label_0.json", 'r', encoding='utf-8') as f:
    label_0_data = json.load(f)
    print("Power Label 0:")
    for chapter_id, sentences in label_0_data.items():
        for sentence in sentences:
            print(f"Chapter {chapter_id}:\n {sentence['sentence']}")
            print()
```

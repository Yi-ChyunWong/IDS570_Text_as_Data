```python
import stanza

nlp = stanza.Pipeline(
    lang='ar',
    dir='C:/Users/yichy/stanza_resources',
    download_method=None,
    use_gpu=True   
)
```

```python
import re

def remove_diacritics_set(word_set):
    return set(re.sub(r'[\u0610-\u061A\u064B-\u065F\u0670]', '', word) for word in word_set)

def open_txt(in_path):
    with open(in_path, 'r', encoding='utf-8') as f:
        text = f.read()
    return text

def save_txt(text, out_path):
    with open(out_path, 'w', encoding='utf-8') as f:
        f.write(text)

def remove_diacritics_txt(text):
    return re.sub(r'[\u0610-\u061A\u064B-\u065F\u0670]', '', text)

def lemmatize_txt(text):
    words = text.split()
    chunk_size = 1000
    chunks = [" ".join(words[i:i+chunk_size]) for i in range(0, len(words), chunk_size)]
    
    lemma = []
    for chunk in chunks:
        doc = nlp(chunk)
        lemma += [
            remove_diacritics_txt(word.lemma)
            for sentence in doc.sentences
            for word in sentence.words
        ]
    return lemma
```

```python
import pickle
import nltk
# nltk.download()
from nltk.corpus import stopwords

def get_arabic_stopwords():
    arabic_stopwords = set(stopwords.words('arabic'))

    # Add punctuation and extra characters
    extra = {
        '،',  # Arabic comma
        '.',  # Period
        ',',  # English comma
        '؟',  # Arabic question mark
        '!',
        ':',
        ';',
        '(',
        ')',
        '-',
        '–',  # en dash
        '—',  # em dash
        '"',
        "'",
        '«',  # Arabic quotation marks
        '»',
        '/',
        '\\',
        '\n',
        '\t',
    }
    arabic_stopwords.update(extra)
    # print(f"Number of Arabic stopwords: {len(arabic_stopwords)}")
    # print(f"Sample Arabic stopwords: {list(arabic_stopwords)[:100]}")
    return arabic_stopwords

def save_stopwords():
    stopwords_set = get_arabic_stopwords()
    stopwords_set_no_diacritics = remove_diacritics_set(stopwords_set)
    stopwords_set.update(stopwords_set_no_diacritics)
    print(f"Total Arabic stopwords (with and without diacritics): {len(stopwords_set)}")
    print(f"Sample Arabic stopwords (with and without diacritics): {list(stopwords_set)[:100]}")

    with open('stopwords/arabic_stopwords.pkl', 'wb') as f:
        pickle.dump(stopwords_set, f)
    
    with open("stopwords/arabic_stopwords.txt", "w", encoding="utf-8") as f:
        for word in sorted(stopwords_set):
            f.write(word + "\n")

save_stopwords()
```

```python
def remove_diacritics_corpus():
    for i in range (0, 130):
        txt = open_txt(f"corpus_txt/chapter_{i}.txt")
        txt_no_diacritics = remove_diacritics_txt(txt)

        save_txt(txt_no_diacritics, f"corpus_txt_no_diacritics/chapter_{i}.txt")

remove_diacritics_corpus()
```

```python
def remove_stopwords_corpus():
    with open('stopwords/arabic_stopwords.pkl', 'rb') as f:
        stopwords = pickle.load(f)
    
    for i in range (0, 130):
        txt = open_txt(f"corpus_txt_no_diacritics/chapter_{i}.txt")
        words = txt.split()
        words_no_stopwords = [word for word in words if word not in stopwords]
        txt_no_stopwords = " ".join(words_no_stopwords)

        save_txt(txt_no_stopwords, f"corpus_txt_no_stopwords/chapter_{i}.txt")

remove_stopwords_corpus()

def remove_stopwords_lemma(lemma_list):
    with open('stopwords/arabic_stopwords.pkl', 'rb') as f:
        stopwords = pickle.load(f)
    
    lemma_no_stopwords = [lemma for lemma in lemma_list if lemma not in stopwords]
    return lemma_no_stopwords
```

```python
import json

def lemmatize_corpus():
    lemmatized_corpus = {}

    for i in range (0, 130):
        txt = open_txt(f"corpus_txt_no_stopwords/chapter_{i}.txt")
        lemmatized_words = lemmatize_txt(txt)
        lemmatized_words_no_stopwords = remove_stopwords_lemma(lemmatized_words)
        lemmatized_corpus[f"chapter_{i}"] = lemmatized_words_no_stopwords
        print(f"Lemmatized chapter_{i}")
    
    with open("corpus_lemmatized.json", "w", encoding="utf-8") as f:
        json.dump(lemmatized_corpus, f, ensure_ascii=False, indent=4)

lemmatize_corpus()
```

```python
import json
import re

def split_corpus_in_sentences():
    sentences_corpus = {}

    for i in range (0, 130):
        txt = open_txt(f"corpus_txt_no_diacritics/chapter_{i}.txt")
        sentences = re.split(r'[.!؟]', txt)
        sentences = [s.strip() for s in sentences if s.strip()]

        chapter_data = []

        for sentence in sentences:
            lemmas = lemmatize_txt(sentence)
            lemmas = remove_stopwords_lemma(lemmas)
            chapter_data.append({
                "sentence": sentence,
                "lemmas": lemmas
            })
        sentences_corpus[f"chapter_{i}"] = chapter_data        
    
    with open("corpus_sentences.json", "w", encoding="utf-8") as f:
        json.dump(sentences_corpus, f, ensure_ascii=False, indent=4)
split_corpus_in_sentences()
   
```
